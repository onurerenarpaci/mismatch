```python
def fib(n):
    a = 1
    b = 1
    c = 0
    for i in range(n-2):
        c = a+b
        a = b
        b = c

    return c
```

```python
import random

def simulate(p):
    a = [False for x in range(p)]
    a.extend([True for x in range(100-p)])
    random.shuffle(a)
    result = []

    while len(a) > 0:
        first = a.pop(0)
        result.append(first)
        if first:
            random.shuffle(a)
            continue
        else:
            b = a.copy()
            random.shuffle(b)
            reveal = True
            for i in range(len(a)):
                if not (a[i] or b[i]):
                    continue
                elif a[i]:
                    reveal = False
                    break
                else:
                    break

            if reveal: a = b.copy()
    return result

normal = [0 for i in range(100)]
rigged = [0 for i in range(100)]
p = 30
for i in range(10000):
    normal_res = [False for x in range(p)]
    normal_res.extend([True for x in range(100-p)])
    random.shuffle(normal_res)

    rigged_res = simulate(p)

    for i in range(99, -1, -1):
        if normal_res[i]:
            break
        else:
            normal[i] += 1

    for i in range(99, -1, -1):
        if rigged_res[i]:
            break
        else:
            rigged[i] += 1
    
```
